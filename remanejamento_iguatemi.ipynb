```python
import pandas as pd

caminho = r"C:\Users\geren\Downloads\data (12).xlsx"
estoque = pd.read_excel(caminho)

estoque 
```

```python
# excluir as ultimas 3 linhas
estoque = estoque.drop(estoque.tail(3).index)

estoque
```

```python
estoque.info()
```

```python
# remanejamento com as colunas: 0, 1 2 3 5 6 8
estoque = estoque.iloc[:, [0, 1, 2, 3, 5, 6, 8]]

estoque
```

```python
# ESTOQUE Matriz	ESTOQUE Iguatemi	ESTOQUE MBS são inteiros
estoque['ESTOQUE Matriz'] = estoque['ESTOQUE Matriz'].astype(int)
estoque['ESTOQUE Iguatemi'] = estoque['ESTOQUE Iguatemi'].astype(int)
estoque['ESTOQUE MBS'] = estoque['ESTOQUE MBS'].astype(int)

estoque.info()
```

```python
estoque
```

```python
# filtrar ESTOQUE Matriz igual a 0
remanejamento = estoque[estoque['ESTOQUE Matriz'] == 0]

```

```python
# exclui a coluna ESTOQUE Matriz
remanejamento = remanejamento.drop(columns='ESTOQUE Matriz')

remanejamento
```

```python
# filtrar linhas onde a coluna ESTOQUE Iguatemi tenha 2 unidades ou mais que a coluna ESTOQUE MBS
remanejamento = remanejamento[remanejamento['ESTOQUE Iguatemi'] >= remanejamento['ESTOQUE MBS'] + 2]

remanejamento
```

```python
# filtrar ESTOQUE MBS igual a 0
remanejamento = remanejamento[remanejamento['ESTOQUE MBS'] == 0]

remanejamento
```

```python
# renomear as colunas
remanejamento.columns = ['produto_codigo', 'produto_descricao', 'produto_cor', 'produto_tamanho', 'estoque_iguatemi', 'estoque_morumbi'] 

remanejamento
```

```python
# salvar o arquivo remanejamento_iguatemi.xlsx 
remanejamento.to_excel('remanejamento_iguatemi.xlsx', index=False)
```
